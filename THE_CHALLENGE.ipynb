```python
df_crypto2=df_crypto[(df_crypto['IsTrading']==True)]
df_crypto2.head()
```

```python
count_row = df_crypto2.shape[0]
count_row
```

```python
df_crypto2.drop(columns=["IsTrading"], inplace=True)
df_crypto2.head()
```

```python
df_crypto3 = df_crypto2.dropna()
```

```python
count_row3 = df_crypto3.shape[0]
count_row3
```

```python
df_crypto4=df_crypto3[(df_crypto3['TotalCoinsMined']>0)]
df_crypto4.head()
```

```python
count_row4 = df_crypto4.shape[0]
count_row4
```

```python
crypto_name = df_crypto4[['CoinName']]
crypto_name.head()
```

```python
df_crypto4.drop(columns=["CoinName"], inplace=True)
df_crypto4.head()
```

```python
df_crypto_num = pd.get_dummies(df_crypto4, columns=["Algorithm", "ProofType"])
df_crypto_num.head()
```

```python
df_crypto_scaled = StandardScaler().fit_transform(df_crypto_num)
print(df_crypto_scaled[0:5])
```

```python
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(df_crypto_scaled)

df_crypto_pca = pd.DataFrame(data = crypto_pca, columns=["PC1", "PC2", "PC3"])
df_crypto_pca.head()
```
